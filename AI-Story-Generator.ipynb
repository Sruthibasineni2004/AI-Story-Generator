```python
# Central function to make requests to Gemini API
def make_gemini_request(endpoint, payload):

    # Make a POST request to the Gemini API.
    # :param endpoint: API endpoint (e.g., 'models/gemini-1.5-flash-latest:generateContent')
    # :param payload: Payload for the request
    # :return: Response JSON or error message
    url = f"https://generativelanguage.googleapis.com/v1beta/{endpoint}?key={gemini_api_key}"
    headers = {"Content-Type": "application/json"}
    try:
        response = requests.post(url, json=payload, headers=headers)
        response.raise_for_status()
        return response.json()
    except requests.RequestException as e:
        return {"error": f"Error with Gemini API: {e}"}
```

```python
#Login function
def login_page():
    clear_output()

    login_username = widgets.Text(description="Username:")
    login_password = widgets.Password(description="Password:")
    login_button = widgets.Button(description="Login")
    signup_redirect_button = widgets.Button(description="Sign Up")

    def on_login_button_clicked(b):
        username = login_username.value
        password = login_password.value
        if username in user_database and user_database[username] == password:
            print("Login successful!")
            home_page(username)
        else:
            print("Invalid username or password.")

    def on_signup_redirect_button_clicked(b):
        signup_page()

    login_button.on_click(on_login_button_clicked)
    signup_redirect_button.on_click(on_signup_redirect_button_clicked)

    display(widgets.VBox([widgets.Label("Login Form"), login_username, login_password, login_button, signup_redirect_button]))
```

```python
# Signup function
def signup_page():
    clear_output()

    signup_username = widgets.Text(description="Username:")
    signup_email = widgets.Text(description="Email:")
    signup_password = widgets.Password(description="Password:")
    signup_button = widgets.Button(description="Sign Up")
    login_redirect_button = widgets.Button(description="Back to Login")

    def on_signup_button_clicked(b):
        username = signup_username.value
        email = signup_email.value
        password = signup_password.value
        if username in user_database:
            print("Username already exists.")
        else:
            user_database[username] = password
            print("Account created successfully!")
            login_page()

    def on_login_redirect_button_clicked(b):
        login_page()

    signup_button.on_click(on_signup_button_clicked)
    login_redirect_button.on_click(on_login_redirect_button_clicked)

    display(widgets.VBox([widgets.Label("Sign Up Form"), signup_username, signup_email, signup_password, signup_button, login_redirect_button]))
```

```python
# Function to generate text with Gemini API
# Function to generate text with Gemini API
def generate_text_with_gemini(prompt):
    payload = {"contents": [{"parts": [{"text": prompt}]}]}
    response = make_gemini_request("models/gemini-1.5-flash-latest:generateContent", payload)

    # Debug: print the full response to verify the structure
    print("API Response:", response)

    # Check if there's an error in the response
    if "error" in response:
        return response["error"]

    try:
        # Access the correct fields in the response based on the provided structure
        return response.get("candidates", [{}])[0].get("content", {}).get("parts", [{}])[0].get("text", "No text generated")
    except Exception as e:
        return f"Error extracting content: {e}"
```

```python
def generate_image_with_gemini(prompt):
    payload = {"prompt": prompt, "n": 1, "size": "512x512"}
    response = make_gemini_request("models/gemini-1.5-flash-latest:generateImage", payload)

    # Debug: print the full response to verify the structure
    print("API Image Response:", response)

    if "error" in response:
        return response["error"]

    try:
        # Get the image URL from the response
        return response.get("images", [{}])[0].get("url", "No image generated")
    except Exception as e:
        return f"Error extracting image URL: {e}"
```

```python
# Home Page with question input
def home_page(username):
    clear_output()

    print(f"Welcome, {username}!")
    question_input = widgets.Text(description="Question:")
    submit_button = widgets.Button(description="Submit")

    def on_submit_button_clicked(b):
        question = question_input.value
        if question:
            # Generate text with Gemini API
            story_text = generate_text_with_gemini(question)
            print("\nGenerated Text:", story_text)

            # Generate an image with Gemini API
            image_url = generate_image_with_gemini(question)
            print("Generated Image URL:", image_url)
        else:
            print("Please enter a question.")

    submit_button.on_click(on_submit_button_clicked)

    display(widgets.VBox([widgets.Label("Storytelling App"), question_input, submit_button]))
```
